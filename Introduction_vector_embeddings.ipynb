```python
import numpy as np

# Example vector
vector = np.array([1, 2, 3])
print("Vector:", vector)
```

```python
from sentence_transformers import SentenceTransformer

# Initialize the model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example sentence
sentence = "boy"

# Generate embedding (vector) for the sentence
embedding = model.encode(sentence)
print("Embedding for 'apple':", embedding)
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
# Step 1: Load documents from the "data" folder
documents = SimpleDirectoryReader("data").load_data()
#print(documents)
#Document = Node 1 + Node 2
# Step 2: Create a vector index from the documents
index = VectorStoreIndex.from_documents(documents)

#print(index)

query_engine = index.as_query_engine()
response = query_engine.query("what is a vector database")
print(response)
```
